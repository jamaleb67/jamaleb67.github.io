## 2. Formulate the request

URLs are addresses to information that is prettied up into a page. We want the data so in a sense APIs tell us how to get the data directly.  

**STOP** Actually go an look at the documentation an [API request](https://datahelpdesk.worldbank.org/knowledgebase/articles/898581-api-basic-call-structures)


````python
address =  'http://api.worldbank.org/V2/country?incomeLevel=LIC' #Stores address in an object named address
resp = requests.get(address) #uses the request library
resp # Shows you what returned
````

````python
#Per the documentation we can add an argument to gives us the format in json so we are going to add an argument "&format=json" to the end
address =  'http://api.worldbank.org/V2/country?incomeLevel=LIC&format=json' #Stores address in an object named address
resp = requests.get(address) #uses the request library
resp.json() # Shows you what returned
````

````python
#but we want to store that data structure in a object so we are not calling the json method each time
LIC = resp.json()
LIC
````
